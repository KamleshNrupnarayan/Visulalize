```python
from matplotlib import pyplot as p
import numpy as np
```

```python
x=np.array([10,20])
y=np.array([20,40])
```

```python
p.plot(x,y)
```

```python
p.plot(x,y,marker="o",)
```

```python
a=np.array([10,30,50,60])
b=np.array([100,200,300,400])
```

```python
p.plot(b,"o")
```

```python
p.plot(b,)
```

```python
p.plot(b,a)
```

```python
p.plot(x,y,'P--m')
```

```python
p.plot(x,y,'P--r',ms=10)
```

```python
p.plot(x,y,'*--r',ms=10,mfc='k')
```
