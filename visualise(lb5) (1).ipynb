```python
def binning(data):
    maximum_value=max(data)
    minimum_value=min(data)
    print("Maximum_age:",maximum_value)
    print("Minimum_age:",minimum_value)

    child=[] #empty list
    teen=[]
    adult=[]
    old=[]
    
    #for i in range(0,len(data)):
    for i in data:
        if i<=12:
            child.append(i)
        elif i>12 and i<20:
            teen.append(i)
        elif i>=20 and i<45:
            adult.append(i)
        elif i>=45:
            old.append(i)

    
    print("bin of children :",child) 
    print("bin of teen:",teen)
    print("bin of old people:",old)
    print("bin of adult:",adult)
age=[10,2,14,23,12,46,67,30,20]
binning(age)
```
